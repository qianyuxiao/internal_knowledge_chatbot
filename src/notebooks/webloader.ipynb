# pdf loader

```python
scanned_pdf_path =  "../documents/ecg_assurance/Contrats 2024/SIEGES/Generali - Si\u00e8ge Montpellier.pdf"
normal_pdf_path = "../documents/ecg_assurance/Contrats 2024/CAMPINGS/Avenant 2024 Police campings Allianz.pdf"
```

### - PyPDFLoader

```python
from langchain_community.document_loaders import PyPDFLoader
```

```python
for pdf_path in [scanned_pdf_path,normal_pdf_path]:
    loader = PyPDFLoader(pdf_path)
    docs = loader.load()
    all_empty = all(not doc.page_content for doc in docs)
    print(all_empty)
```

### Google loader

```python
from langchain.document_loaders import GCSDirectoryLoader
```

```python
PROJECT_ID = 'ecg-ai-416210'
```

```python
# normal pdf
GCS_BUCKET_DOCS = 'ecg_hr'
loader = GCSDirectoryLoader(
    project_name=PROJECT_ID, bucket=GCS_BUCKET_DOCS, prefix="original_documents")
pages = loader.load()
print(pages[0].page_content)
```

```python
# Add document name and source to the metadata
for document in pages:
    doc_md = document.metadata
    document_name = doc_md["source"].split("/")[-1]
    # derive doc source from Document loader
    doc_source_prefix = "/".join(GCS_BUCKET_DOCS.split("/")[:3])
    doc_source_suffix = "/".join(doc_md["source"].split("/")[4:-1])
    source = f"{doc_source_prefix}/{doc_source_suffix}"
    document.metadata = {"source": source, "document_name": document_name}
```

```python
print(pages[0].page_content[:1000])
```

```python
print(pages[1].page_content[:1000])
```

```python
print(pages[1].page_content[1000:5000])
```
